```
def formata_coluna(coluna):
 
  if coluna[-3:] == 'Jan':
    coluna = coluna.replace('Jan', '01')
  elif coluna[-3:] == 'Fev':
    coluna = coluna.replace('Fev', '02')
  elif coluna[-3:] == 'Mar':
    coluna = coluna.replace('Mar', '03')
  elif coluna[-3:] == 'Abr':
    coluna = coluna.replace('Abr', '04')
  elif coluna[-3:] == 'Mai':
    coluna = coluna.replace('Mai', '05')
  elif coluna[-3:] == 'Jun':
    coluna = coluna.replace('Jun', '06')
  elif coluna[-3:] == 'Jul':
    coluna = coluna.replace('Jul', '07')
  elif coluna[-3:] == 'Ago':
    coluna = coluna.replace('Ago', '08')
  elif coluna[-3:] == 'Set':
    coluna = coluna.replace('Set', '09')
  elif coluna[-3:] == 'Out':
    coluna = coluna.replace('Out', '10')
  elif coluna[-3:] == 'Nov':
    coluna = coluna.replace('Nov', '11')
  elif coluna[-3:] == 'Dez':
    coluna = coluna.replace('Dez', '12')
  
  coluna = coluna.replace('/','')
  return coluna
```
